```python
u1 = np.array([0.1,0.5])
cov1  = np.array([[1,-0.8],
                  [-0.8,2]])

u2 = np.array([5.1,2.5])
cov2  = np.array([[1,0.5],
                  [0.5,1]])


data1 = np.random.multivariate_normal(u1,cov1,1000)
data2 = np.random.multivariate_normal(u2,cov2,1000)
```

```python
print(data1.shape)
X1 = data1
X2 = data2
```

```python
import matplotlib.pyplot as plt
plt.scatter(X1[:,0],X1[:,1])
plt.scatter(X2[:,0],X2[:,1],color='red')
plt.show()
```

```python
X = np.vstack((X1,X2))
print(X.shape)
```

```python
Y = np.zeros((X.shape[0],))
Y[1000: ] = 1
```

```python
print(Y)
print(Y.shape)
```

```python
styles = plt.style.available
print(styles)
```

```python

plt.style.use('seaborn')
plt.figure(figsize=(10,6))
plt.title("Data")
plt.xlabel("X1")
plt.ylabel("X2")
plt.scatter(X[:,0],X[:,1],c=Y,cmap=plt.cm.Accent)
plt.show()
```

```python
final_data = np.hstack((X,Y.reshape(-1,1)))
print(final_data.shape)
```

```python
np.save("Data.npy",final_data)
```
