```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.load('LinearX.npy')
Y = np.load('LinearY.npy')
```

```python
X = (X-X.mean())/X.std()
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
def y_hat(theta,x):
    return theta[0] + theta[1]*x

def error(X,Y,theta):
    Y_ = y_hat(theta,X)
    e =  np.sum((Y_ - Y)**2)
    m = X.shape[0]
    return e/m

def getGrad(X,Y,theta):
    grad = np.zeros((2,))
    Y_ = y_hat(theta,X)
    
    grad[0] = 2*np.sum(Y_ - Y)
    grad[1] = 2*np.sum( (Y_ - Y)*X)
    m = X.shape[0]
    return grad/m
    

def gradDescent(X,Y,lr=0.01):
    theta = np.zeros((2,))
    
    error_list = []
    for i in range(200):
        grad = getGrad(X,Y,theta)
        theta = theta - lr*grad
        e = error(X,Y,theta)
        error_list.append(e)
        
    return theta,error_list
    
```

```python
theta,error_list = gradDescent(X,Y)
```

```python
plt.plot(error_list)
```
