```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
print(X)
```

```python
# Let us assume some fn of X
Y = (X-5)**2
print(Y)
```

```python
plt.plot(X,Y)
plt.show()
```

```python
# Gradient Descent
plt.plot(X,Y)

x_ = 1
lr  = 0.1
error = []
for i in range(50):
    dy_dx = 2*(x_-5)
    x_ = x_ - lr*dy_dx
    y = (x_-5)**2
    
    print("X : %.2f Error %.2f"%(x_,y))
    error.append(y)
    plt.scatter(x_,y)
    

plt.show()
```

```python
plt.plot(error)
plt.title("Error vs Iterations")
plt.show()
```
