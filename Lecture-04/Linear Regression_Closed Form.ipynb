```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.load('LinearX.npy')
Y = np.load('LinearY.npy')
```

```python
X = (X-X.mean())/X.std()
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
X = X.reshape((-1,1))
ones = np.ones((X.shape[0],1))
X_new = np.hstack((ones,X))
print(X_new[:5,:])
```

```python
X = X_new
```

```python
theta = np.dot(np.linalg.pinv(np.dot(X.T,X)),np.dot(X.T,Y))
print(theta)
```

```python
plt.scatter(X[:,1],Y)
Y_ = np.dot(X,theta)
plt.plot(X[:,1],Y_,color='green')
plt.show()
```
